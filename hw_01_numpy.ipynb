# Задача 1

## Условие
Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы. 

```python
test1 = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])
test2 = np.arange(-100, 100, 2).reshape(10, -1)
test3 = np.random.randn(33, 33)
print("Размеры массивов:")
print(f"test1 {test1.size}")
print(f"test3 {test2.size}")
print(f"test3 {test3.size}")
```

## 1.1

```python
def solv_1_1(X):
  Y = np.diag(X)
  return np.prod(Y[Y != 0])
solv_1_1(test1)
```

## 1.2

```python
def solv_1_2(X):
  res = 1;
  for i in np.arange(min(X.shape)):
    if X[i, i] != 0:
      res *= X[i, i]
  return res;
solv_1_2(test1)
```

## 1.3

```python
def solv_1_3(X):
  Y = np.diag(X)
  res = 1
  for i in Y:
    if i != 0:
      res *= i
  return res
solv_1_3(test1)
```

## Сравнение результатов

```python
%timeit -n 1000 solv_1_1(test1)
%timeit -n 1000 solv_1_1(test2)
%timeit -n 1000 solv_1_1(test3)
```

```python
%timeit -n 1000 solv_1_2(test1)
%timeit -n 1000 solv_1_2(test2)
%timeit -n 1000 solv_1_2(test3)
```

```python
%timeit -n 1000 solv_1_3(test1)
%timeit -n 1000 solv_1_3(test2)
%timeit -n 1000 solv_1_3(test3)
```

## Условие
Даны два вектора x и y. Проверить, задают ли они одно и то же мультимножество. 

```python
x_test_3_1 = np.array([1, 2, 2, 4])
y_test_3_1 = np.array([4, 2, 1, 2])

x_test_3_2 = np.arange(40)
y_test_3_2 = np.array(x_test_3_2[::-1])

x_test_3_3 = np.arange(400)
y_test_3_3 = np.array(x_test_3_3[::-1])
```

## 3.1

```python
def solv_3_1(X, Y):
  if (X.size != Y.size):
    return False
  Z = np.array(np.sort(X) == np.sort(Y), dtype=int)
  return Z.size == Z.sum()
solv_3_1(x_test_3_1, y_test_3_1)
```

## 3.2

```python
def solv_3_2(X, Y):
  if (len(X) != len(Y)):
    return False
  for i in range(len(X)):
    if X[X.argmin()] != Y[Y.argmin()]:
      return False
    np.delete(X, X.argmin())
    np.delete(Y, Y.argmin())
  return True
solv_3_2(x_test_3_1, y_test_3_1)
```

## 3.3

```python
def solv_3_3(X, Y):
  if (X.size != Y.size):
    return False
  X.sort()
  Y.sort()
  for i in range(X.size):
    if X[i] != Y[i]:
      return False
  return True
solv_3_3(x_test_3_1, y_test_3_1)
```

## 4.1

Функция работает по принципу сдвига: сдвигаем истинные значение при х == 0 для "урезанного" начального массива и находим соответствующие пересечения, находим максимум среди удовлетворяющих значений

```python
def solv_4_1(X):
  return X[1:][(X == 0)[:-1]].max()
```

## 4.2

```python
def solv_4_2(X):
  lst = list()
  for i in range(1, X.size):
    if X[i - 1] == 0:
      lst.append(X[i])
  return max(lst)
```

## 4.3

```python
def solv_4_3(X):
  z = (X == 0)[:-1]
  max = -1
  for i in np.arange(X.size - 1):
    buf = ~X[1:][i].astype(bool) == z[i]  # both elem equal NULL
    if buf == True & max < X[i]:
      max = X[i]
      buf = False
  return max
```
